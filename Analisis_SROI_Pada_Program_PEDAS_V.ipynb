```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

def label_encode_columns(dataset, columns_to_encode):
    label_encoders = {}  # Dictionary to store LabelEncoder objects
    mappings = {}  # Dictionary to store mappings

    for column in columns_to_encode:
        if column in dataset.columns:
            if dataset[column].dtype == 'object' or dataset[column].dtype == 'float64' or dataset[column].dtype == 'bool':
                label_encoder = LabelEncoder()
                dataset[column] = label_encoder.fit_transform(dataset[column])
                label_encoders[column] = label_encoder
                mappings[column] = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

    return dataset, mappings

# List the columns you want to label encode
columns_to_encode = ['Jenis Kelamin', 'Pendapatan Tambahan ',
       'Besar Pendapatan tambahan', 'Besar Pengurangan Biaya OP',
       'Besar Penghematan Biaya PP', 'Pengurangan Kerusakan Lingkungan',
       'Jenis Kerusakan Lingkungan  ', 'Meningkatkan Pengetahuan ',
       'Tingkat Perubahan Positif ', 'Manfaat Utama ', 'Tingkat Kepuasan']

# Assuming df is your DataFrame
df_encoded, mappings = label_encode_columns(df, columns_to_encode)

# Display the mappings
for column, mapping in mappings.items():
    print(f"Mapping for column {column}:")
    for key, value in mapping.items():
        print(f"{key}: {value}")
    print("\n")

# To see the encoded DataFrame
print(df_encoded.head())
```

```python
from scipy.stats import chi2_contingency

def calculate_phi_correlation(df, col1, col2):
  contingency_table = pd.crosstab(df[col1], df[col2])
  chi2, p, _, _ = chi2_contingency(contingency_table)
  n = contingency_table.sum().sum()
  phi = np.sqrt(chi2 / n)
  return phi

# Calculate phi correlation for the specified pairs
columns_to_correlate = ['Pendapatan Tambahan ', 'Pengurangan Kerusakan Lingkungan', 'Tingkat Kepuasan', 'Meningkatkan Pengetahuan ']

for i in range(len(columns_to_correlate)):
  for j in range(i + 1, len(columns_to_correlate)):
    col1 = columns_to_correlate[i]
    col2 = columns_to_correlate[j]
    phi_correlation = calculate_phi_correlation(df_encoded, col1, col2)
    print(f"Phi correlation between {col1} and {col2}: {phi_correlation}")
```

## Perhitungan SROI

```python
# Menghitung frekuensi untuk kolom-kolom tertentu
columns_to_analyze = ['Besar Pendapatan tambahan', 'Besar Pengurangan Biaya OP',
       'Besar Penghematan Biaya PP']

frequency_counts = {}
for column in columns_to_analyze:
    frequency_counts[column] = df[column].value_counts()

# Menampilkan hasil frekuensi
for column, counts in frequency_counts.items():
    print(f"Frekuensi untuk {column}:")
    print(counts)
    print("\n")
```

```python
mapping_Besar_Pendapatan_tambahan = {
    0: 250000,
    1: 750000
}

mapping_Besar_Pengurangan_Biaya_OP = {
    0: 50000,
    1: 200000
}

mapping_Besar_Penghematan_Biaya_PP = {
    0: 50000,
    1: 200000
}
```

```python
# Fungsi untuk menghitung rata-rata tertimbang
def calculate_weighted_average(df, column, mapping):
    # Hitung frekuensi untuk setiap kategori
    frequency_counts = df[column].value_counts()

    # Hitung nilai total berdasarkan frekuensi dan mapping
    total_value = 0
    total_frequency = 0
    for category, freq in frequency_counts.items():
        total_value += mapping[category] * freq
        total_frequency += freq

    # Hitung rata-rata tertimbang
    if total_frequency == 0:
        return 0
    weighted_average = total_value / total_frequency
    return weighted_average

# Hitung rata-rata untuk setiap kolom
average_Besar_Pendapatan_tambahan = calculate_weighted_average(df, 'Besar Pendapatan tambahan', mapping_Besar_Pendapatan_tambahan)
average_Besar_Pengurangan_Biaya_OP = calculate_weighted_average(df, 'Besar Pengurangan Biaya OP', mapping_Besar_Pengurangan_Biaya_OP)
average_Besar_Penghematan_Biaya_PP = calculate_weighted_average(df, 'Besar Penghematan Biaya PP', mapping_Besar_Penghematan_Biaya_PP)

# Tampilkan hasil
print(f"Rata-rata Pendapatan Tambahan: Rp {average_Besar_Pendapatan_tambahan:.2f}")
print(f"Rata-rata Pengurangan Biaya Operasioanal : Rp {average_Besar_Pengurangan_Biaya_OP:.2f}")
print(f"Rata-rata Penghematan Biaya Pupuk dan Pestisida: Rp {average_Besar_Penghematan_Biaya_PP:.2f}")
```

```python
# Hitung total nilai ekonomi untuk setiap kolom
# Hitung nilai ekonomi total
# Nilai ekonomi = indikator * proksi - deadweight - attribution - drop off
nilai_ekonomi_total_besar_pendapatan_tambahan = 35 * average_Besar_Pendapatan_tambahan - 0 - 0 - 0
nilai_ekonomi_total_besar_pengurangan_biaya = 35 * average_Besar_Pengurangan_Biaya_OP - 0 - 0 - 0
nilai_ekonomi_total_besar_penghematan_biaya = 35 * average_Besar_Penghematan_Biaya_PP - 0

# Tampilkan hasil
print(f"Nilai Dampak Besar Pendapatan tambahan: Rp {nilai_ekonomi_total_besar_pendapatan_tambahan:.2f}")
print(f"Nilai Dampak Besar Pengurangan Biaya Operasiona;: Rp {nilai_ekonomi_total_besar_pengurangan_biaya:.2f}")
print(f"Nilai Dampak Besar Besar Penghematan Biaya Pupuk & Pestisida: Rp {nilai_ekonomi_total_besar_penghematan_biaya:.2f}")
```

```python
# Hitung nilai SROI

outcome = (nilai_ekonomi_total_besar_pendapatan_tambahan + nilai_ekonomi_total_besar_pengurangan_biaya + nilai_ekonomi_total_besar_penghematan_biaya)

outcome_tahunan = outcome*12
total_biaya_program = 506725025

sroi = outcome_tahunan/ total_biaya_program

print(f'Total Nilai Ekonomi yang Dihasilkan: Rp {outcome}')
print(f'Total nilai ekonomi satu tahun: Rp {outcome_tahunan}')
print(f'Nilai SROI: {sroi:.2f}')
```
